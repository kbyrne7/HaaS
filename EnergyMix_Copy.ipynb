```python
# Energy data to display energy usage by mix of energy resources
from ast import Num
# Load CSV Using Python Standard Library
import csv
import numpy
import requests
import pandas as pd
# Open URL / API
from urllib.request import urlopen

# Pandas and math polot library
from pandas import set_option
from pandas import read_csv
#from matplotlib import pyplot
# Ref the scatter plot matrix
# Use Pandas to read CSV

# Open file
url                  = 'https://api.carbonintensity.org.uk/generation'
names                = ['Timestamp', 'Facility_IT_Load_MW', 'Supply_Temp_C', 'Return_Temp_C', 'Flow_m3h', 'Grid_gCO2_kWh']
# 1. Fetch the data
raw_data             = requests.get(url)
# 2. Convert to dictionary
raw_json             = raw_data.json()
# 3. FIX: Access the data using the STRING 'data' as the key
# The API returns data: {"data": {"from": "...", "to": "...", "generationmix": [...]}}
actual_data             = raw_json['data']
# 4. Flatten the 'generationmix' nested list into a table and normalize the data
dataset                 = pd.json_normalize(actual_data['generationmix'])
# Format date/time col
dataset['Timestamp']    = raw_json['data']['to']
dataset                 = dataset.rename(columns={'perc': 'Percentage'})
# 4. Format the Date (ISO8601 is standard for 2026)
dataset['Timestamp']    = pd.to_datetime(dataset['Timestamp'], format='ISO8601')
dataset['Timestamp']    = dataset['Timestamp'].dt.strftime('%Y-%m-%d %H:%M:%S')
# Sort data by percentage of energy source
dataset             = dataset.sort_values(by='Percentage', ascending=False)
# Reset index for sort
dataset             = dataset.reset_index(drop=True)

print(dataset.shape)
print(dataset.head())

```
